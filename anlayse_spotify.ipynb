# Exploratory Data Analysis (EDA) on Spotify Dataset

This notebook performs an EDA on the Spotify dataset to uncover patterns and insights. We will investigate distributions of key features, analyse trends over time, and explore correlations between various musical attributes and popularity.

## Data Overview

The dataset contains several attributes, including artist name, song title, duration, popularity, and audio features like danceability, energy, and valence. 

In order to understand the distribution and ranges of the attributes, we have ccomputed the following basic statistics

Now we will push the analysis by visualising the distribution of Popularity, Danceability, Energy, Valance and Tempo.
<br><br><br><br>

```python

fig, axes = plt.subplots(3, 2, figsize=(16, 12))
fig.suptitle("Distributions of Key Features in Spotify Data", fontsize=16)

# Distribution of Popularity
sns.histplot(spotify_data['popularity'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribution of Popularity')
axes[0, 0].set_xlabel('Popularity')
axes[0, 0].set_ylabel('Frequency')

# Distribution of Danceability
sns.histplot(spotify_data['danceability'], kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Distribution of Danceability')
axes[0, 1].set_xlabel('Danceability')
axes[0, 1].set_ylabel('Frequency')

# Distribution of Energy
sns.histplot(spotify_data['energy'], kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Distribution of Energy')
axes[1, 0].set_xlabel('Energy')
axes[1, 0].set_ylabel('Frequency')

# Distribution of Valence
sns.histplot(spotify_data['valence'], kde=True, ax=axes[1, 1], color='lightcoral')
axes[1, 1].set_title('Distribution of Valence')
axes[1, 1].set_xlabel('Valence')
axes[1, 1].set_ylabel('Frequency')

# Distribution of Tempo
sns.histplot(spotify_data['tempo'], kde=True, ax=axes[2, 0], color='gold')
axes[2, 0].set_title('Distribution of Tempo')
axes[2, 0].set_xlabel('Tempo (BPM)')
axes[2, 0].set_ylabel('Frequency')


axes[2, 1].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## Distributions of Key Features

The charts shows the distribution of several key features, specifically: popularity, danceability, energy, valence, and tempo. Here is an interpretation of each distribution:

Popularity: The popularity distribution is right-skewed, with a peak around the mid-range (60-80), indicating that most songs in this dataset have moderate popularity scores. A few songs have very low popularity, possibly reflecting niche tracks or less accessible music styles.

Danceability: The danceability distribution shows a strong concentration around 0.6-0.8, suggesting that a majority of the songs are reasonably danceable, but not extremely so. The right-skew of this distribution indicates that high danceability is common, which aligns with the genre's aim to be appealing and rhythm-driven.

Energy: The energy distribution is left-skewed, with most values concentrated around 0.6 to 1.0. This suggests that the songs tend to be energetic, reflecting an upbeat and lively music trend. Lower energy tracks are rare in this dataset.

Valence: Valence, which measures the positivity or happiness of the music, has a near-normal distribution centered around 0.5 to 0.7. This indicates a balanced spread of both positive and neutral emotional tones in the songs. Songs with very low or very high valence are less common, suggesting that highly intense moods (very happy or very sad) are less typical.

Tempo: The tempo distribution has multiple peaks around common BPM values like 120, a standard tempo for popular music, with additional smaller peaks at other BPM ranges. This distribution reflects diverse tempo preferences, although the most common tempo is around 120 BPM, a pace often found in popular and dance music.

Overall, this distribution analysis reveals a dataset dominated by energetic, moderately danceable, and fairly popular songs, with tempo preferences centered around typical BPM values for pop music. This gives insight into the general characteristics that might appeal to a broad audience in this genre.

```python
# Plot trends of popularity, energy, and danceability over the years
fig, ax = plt.subplots(3, 1, figsize=(14, 12), sharex=True)
fig.suptitle("Trends of Popularity, Energy, and Danceability Over the Years", fontsize=16)

# Average Popularity Over the Years
sns.lineplot(data=spotify_data, x='year', y='popularity', ax=ax[0], marker='o', color='steelblue')
ax[0].set_title('Average Popularity Over the Years')
ax[0].set_ylabel('Popularity')

# Average Energy Over the Years
sns.lineplot(data=spotify_data, x='year', y='energy', ax=ax[1], marker='o', color='seagreen')
ax[1].set_title('Average Energy Over the Years')
ax[1].set_ylabel('Energy')

# Average Danceability Over the Years
sns.lineplot(data=spotify_data, x='year', y='danceability', ax=ax[2], marker='o', color='coral')
ax[2].set_title('Average Danceability Over the Years')
ax[2].set_ylabel('Danceability')
ax[2].set_xlabel('Year')

# Format x-axis as integer only
ax[2].xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## Trends Over the Years

Popularity:
The popularity of songs has remained relatively stable from the late 1990s until around 2018, with an average score around 60-65.
There is a noticeable spike in popularity in the most recent years, suggesting that newer songs in the dataset might be gaining more traction, possibly due to increased accessibility through streaming platforms.
The drop in popularity at the very end may reflect an incomplete dataset for the latest years or simply fewer data points.

Energy:
Energy levels in songs have remained fairly constant since the early 2000s, hovering around an average of 0.7 to 0.8, indicating that most songs are consistently energetic.
The slight dip in recent years may suggest a trend toward slightly less intense or high-energy songs, though this change is minimal.

Danceability:
Danceability has shown a subtle upward trend, with a gradual increase over the years.
Starting from a lower baseline in the late 1990s, danceability has risen steadily, indicating a shift towards songs that are more rhythm-driven and suitable for dancing.
The upward trend aligns with an increasing preference for music that is more groove-oriented, perhaps influenced by the rise of genres like pop, hip-hop, and electronic dance music.

## Feature Importance for Song Popularity Prediction

Valence: Valence, which represents the positivity or happiness of a song, has the highest importance in predicting popularity. This suggests that songs with a more positive or happy tone tend to be more popular, making valence a key factor.

Loudness: Loudness, which indicates the overall volume, ranks as the second most important feature. Louder songs are often perceived as more intense and engaging, which might contribute to their popularity.

Energy: Energy, closely related to loudness, is also an important factor. Higher-energy songs, often associated with excitement and intensity, are more likely to be popular.

Acousticness: Acousticness, which measures how acoustic a track is, has moderate importance. Songs with lower acousticness (i.e., more electronic or synthetic sounds) may appeal more to listeners, but this relationship is less direct than with valence or loudness.

Speechiness: Speechiness, which measures the presence of spoken words, has relatively low importance. This suggests that songs with more spoken content (like rap) are not necessarily more popular.

Instrumentalness: Instrumentalness, which predicts if a track is instrumental, has the least importance. This indicates that the presence of vocals is generally preferred in popular songs.


This feature importance analysis highlights that valence, loudness, and energy are the most critical attributes for predicting popularity, suggesting that upbeat, loud, and energetic songs resonate more with listeners. For artists aiming to create popular music, focusing on a positive tone, high energy, and adequate loudness could increase a song's appeal.
